人工神经网络(Artificial Neural Network，ANN)是模拟大脑中由突触连接起来的大规模神经元集群的数学模型，其可视为一个具有多个联结的计算单元组成的计算系统，能够用于实现非线性计算和模式识别。在众多的ANN模型中，多层感知机（Multiple Layers Perceptron）是一种典型的前馈式多层神经网络，其中，神经元之间是向前连接的[1](#ref1)。在本课程学习报告中，首先总结**多层感知机**模型的基本原理和算法，进而以**非线性动态系统模型辨识**问题作为案例，展示**多层感知机**模型的应用和基于Python程序语言实现的基本步骤。

## 多层感知机模型

一个典型的三层的MLP模型如下图所示[2](#ref2)：

在该ANN模型中，设定：该模型的输入信息为模式向量$\mathbf{u}=\left[u_1,u_2,\ldots,u_n\right]^{\text{T}}$，其对应一个输出向量为$\mathbf{y}=\left[y_1,y_2,\ldots,y_m\right]^{\text{T}}$，此外，向量$\mathbf{v}=\left[v_1,v_2,\ldots,v_p\right]^{\text{T}}$和向量$\mathbf{z}=\left[z_1,z_2,\ldots,z_q\right]^{\text{T}}$分别是第一和第二个中间隐层(hidden layerd)的输出向量。$\left\{w_{i,j}^{1}\right\}_{p\times n}$，$\left\{w_{k,i}^{2}\right\}_{q\times p}$，$\left\{w_{l,k}^{3}\right\}_{m\times q}$输出信号到第一个隐层、第一个隐层到第二个隐层、第二个隐层到第三个隐层之间连接的权值。$\overline{\mathbf{v}}$，$\overline{\mathbf{z}}$，$\overline{\mathbf{y}}$分别是两个中间隐层和输出层神经元的输入信号向量。给定期望的输出信号向量为$\mathbf{y}_{d}=\left[y_{d1},y_{d2},\ldots,y_{dm}\right]^{\text{T}}$，则误差信号可以写为$\mathbf{e}=\mathbf{y}_d-\mathbf{y}$。


## Back Propagation(BP)算法

根据上述示例，在推导多层感知机ANN模型的学习算法时，为了方便，本报告采用如下统一的数学符号：
+ 分别用$i$，$j$，$k$作为ANN模型中不同网络层的神经元的引索；由于信号在前馈式ANN模型中是从左到右传播，因此，设神经元$i$所在的网络层在神经元$j$的左侧，而神经元$j$所在的网络层在神经元$k$的左侧；
+ 用$n$表示输入神经网络模型中的模式信号引索，也可以作为训练模式的引索；
+ 用$E(n)$表示神经网络输出的瞬时的误差平方和；
+ 用$e_j(n)$表示神经元$j$输出的信号的误差；
+ 用$d_j(n)$表示神经元$j$的期望输出信号；
+ 用$y_{j}(n)$表示神经元$j$的实际输出信号；
+ 用$w_{j,i}(n)$表示神经元$i$输出信号输入到神经元$j$的连接权值，对应的矫正量为$\delta w_{j,i}(n)$；
+ 用$v_{j}(n)$表示神经元$j$的输入信号；
+ 用$\phi_j(\cdot)$表示神经元$j$上采用的激励函数；
+ 用$b_j$表示神经元$j$上的偏置值，为了统一表达式，将该偏置值作为一种特殊的连接权重，亦即令$w_{j,0}=b_j$，该连接权重对应的输入信号为常数+1；
+ 用$x_i(n)$表示输入模式信号向量中的第$i$个分量；
+ 用$o_k(n)$表示输出模式信号向量中的第$k$个分量；
+ 用$\eta$表示神经网络学习速率；
+ 用$m_l$表示第$l$层网络的神经元数目，其中，$l=0,1,\ldots,L$，$L$表示神经网络的深度，$m_0$表示输入层网络的神经元数目，其等于输入模式信号向量中分量个数，$m_L$表示输出层网络的神经元个数，其等于输出模式信号向量中分量的个数；中间隐层个数为$L-2$。

训练多层感知机的ANN模型通常可以采用误差反向传播算法，称为**Back Propagation(BP)**算法。该算法通常包括两个过程：一个是信号向前传播过程(forward pass)；第二个是误差反向传播过程(backward pass)。具体如下所述：

### Forward Pass

+ 计算第$l$个网络层的神经元$j_{l}$的输入信号($l=1,2,\ldots,L$)：
$$v_{j_{l}}(n)=\sum_{i_{l-1}=0}^{m_{l-1}}{w_{j_{l},i_{l-1}}(n)y_{i_{l-1}}(n)}$$
其中，$i_{l-1}$表示第$l-1$个网络层的神经元；
+ 计算第$l$个网络层的神经元$j_{l}$的输出信号($l=1,2,\ldots,L$)：
$$y_{j_{l}}(n)=\phi_{j_{l-1}}\left(v_{j_{l}}(n)\right)$$
其中，激励函数统一采用sigmoid函数，亦即，
$$\phi_{j_{l}}(x)=\phi(x)=\frac{1}{1+e^{-\alpha x}}$$
该函数关于$x$的导数可以写为
$$\phi^{\prime}(x)=\alpha \phi(x)\left[1-\phi(x)\right]$$

值得注意的是，当$l=0$时候，输入层神经元的输入信号为$v_{j_0}(n)=x_{i}(n)$，输出信号直接设置为$y_{j_0}(n)=v_{j_0}(n)$。

### Backward Pass

为了调整神经网络模型中权值，在本报告展示中，采用类似随机梯度下降方法，每一个迭代步只考虑处理一个输入模式信号(亦即，仅处理第$n$个模式信号)：首先计算每一个输出层神经元的误差，然后，误差反向传播，一次更新每两个层次之间的权值，这种更新权值的机制称为“序列”调整策略，与之相对应的还有“批量”调整策略。在本报告中，示例程序的实现采用“序列”调整策略，具体流程为：
+ 计算输出层(第$L$层)所有神经元$j_L$的误差：
$$e_{j_L}(n)=d_{j_L}(n)-y_{j_L}(n)$$
+ 计算整个模型瞬时误差：
$$E(n)=\frac{1}{2}\sum_{j_L=1}^{m_L}{e_{j_L}^{2}(n)}$$
整个网络的平均瞬时误差为：
$$E_{avg}=\frac{1}{N}\sum_{n=1}^{N}{E(n)}$$
+ 计算网络的局部梯度$\delta_{j_l}(n)$：
$$
\delta_{j_l}(n)=
\begin{cases}
e_{j_L}(n)\phi^{\prime}_{j_L}\left(v_{j_L}(n)\right),~\text{for }l=L;\\
\phi^{\prime}_{j_l}\left(v_{j_l}(n)\right)\sum_{j_{l+1}=1}^{m_{l+1}}
{\delta_{j_{l+1}}(n)w_{j_{l+1},j_{l}}(n)},~\text{for }l=1,2,\ldots,L-1.
\end{cases}
$$
+ 更新权重：
$$w_{j_{l},j_{l-1}}(n+1)=w_{j_{l},j_{l-1}}(n)+\alpha\left[w_{j_{l},j_{l-1}}(n-1)\right]+\eta\delta_{j_l}y_{j_{l-1}}(n)$$

上述Forward Pass和Backward pass构成一次完整的迭代更新，整个神经网络迭代过程直至性能指标$E_{avg}$小于给定阈值，即可视为收敛到稳定状态。

# 多层感知机在非线性动力系统模型辨识中的应用

为了展示ANN模型的非线性逼近能力，本报告考虑如下非线性动力系统模型的辨识[2](#ref2)：给定离散的时间用$t$表示，$t=0,1,2,\ldots$。一个2阶单输入单输出(single-input-single-output, SISO)非线性动力系统可以表示为
$$
y(t+1)=f\left[y(t),y(t-1)\right]+u(t)
$$
其中，$u(t)$表示时刻$t$的控制输入信号，$y(t)$表示时刻$t$的系统输出信号；$f\left[y(t),y(t-1)\right]$的具体表达式为：
$$
f\left[y(t),y(t-1)\right]=\frac{y(t)y(t-1)\left[y(t)+2.5\right]}
{1+y^2(t)+y^{2}(t-1)}
$$
在设定$f\left[y(t),y(t-1)\right]$未知的情况下，利用神经网络模型实现该非线性动力系统模型辨识的目标即为：构造一个合理的神经网络模型$ANN$，代替该$f\left[y(t),y(t-1)\right]$，使得输出信号：
$$
\widehat{y}(t+1)=ANN\left[y(t),y(t-1)\right]+u(t)
$$
能够很好地逼近实际信号$y(k+1)$。

该非线性动力系统模型的实现代码如下：


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plant(signal, current):
    #the input signal is an numpy-type array
    f = (signal[-1])*(signal[-2])*(signal[-1]+2.5)
    g = 1.0 + signal[-1]**2 + signal[-2]**2
    return f/g + current


def system_model(signal, current, t):
    if t<2:
        inputs = np.array([0.0,0.0])
    else:
        inputs = signal[t-2:t]#[y(t-2),y(t-1)] = inputs
    return plant(inputs,current)
```

本报告给定的控制信号为：$u(t)=sin(2\pi t/25)$，如下：

```python
def control(t):
    if t<0.0:
        return 0.0
    else:
        return np.sin(2*np.pi*t/25.0)
```

实际系统的输出信号波形如下所示(给定时间区间为$[0,100]$)：

```python
epoch = np.linspace(0,100,100)
y = np.zeros((len(epoch),),dtype=float)

for t in range(len(epoch)):
    u_t = control(t-1)
    y[t] = system_model(y,u_t,t)

plt.figure(0,figsize=(8,7))
plt.grid(True)
plt.plot(epoch,y,"-r",lw=4.0,label="$y(t)$")
plt.legend(fontsize=20.0,loc="lower left")
plt.xlabel("Epoch $t$",fontsize=20.0,family="serif")
plt.ylabel("System output",fontsize=20.0,family="serif")
plt.xticks(fontsize=20.0)
plt.yticks(fontsize=20.0)
plt.show()
```

根据示例中动力系统的输入和输出信号维度可知，所要构造的神经网络为2输入1输出的多层感知机模型，具体如下实现：

```python
import copy

class ANN(object):
    
    def __init__(self,structure=(2,20,10,1), eta=0.25, alpha=0.05, TDL=4, a=1.0):
        self.structure = structure
        self.eta = eta
        self.alpha = alpha
        self.a = a
        self.weight = []
        self.new_weight = []
        self.local_gradient = []
        self.local_output = []
        self.output_error = []
        if TDL<1:
            raise("The time delay length (TDL) cannot be less than 1!")
        self.TDL = TDL
        
    
    def Initialization(self,initial_weight_bound = (-1.0,1.0)):
        L = len(self.structure)
        a = initial_weight_bound[0]
        b = initial_weight_bound[-1]
        for l in range(L-1):
            W_l = a+(b-a)*np.random.rand(self.structure[l+1],self.structure[l]+1,self.TDL)
            W_l2 = a+(b-a)*np.random.rand(self.structure[l+1],self.structure[l]+1)
            self.weight.append(W_l)
            self.new_weight.append(W_l2)
            
        for l in range(L-1): 
            if (l<(L-2)):
                self.local_gradient.append(np.zeros((self.structure[l+1]+1,self.TDL),dtype=float))
                self.local_output.append(np.zeros((self.structure[l+1]+1,self.TDL),dtype=float))
            else:
                self.local_gradient.append(np.zeros((self.structure[l+1],self.TDL),dtype=float))
                self.local_output.append(np.zeros((self.structure[l+1],self.TDL),dtype=float))
                
        self.output_error = np.zeros((self.structure[-1],self.TDL),dtype=float)
        return None
    
    def phi(self,x):
        return 1.0/(1.0+np.exp(-1.0*self.a*x))
    
    def dphi(self,x):
        return (self.a)*(self.phi(x))*(1.0-self.phi(x))
    def dphi2(self,y):
        return self.a*y*(1.0-y)
    
    def Forward_Pass(self,patterns,targets,controls,estimations,error_trace):
        L = len(self.structure)
        
        #calculate the outputs of each layer
        for t in range(self.TDL):
            pattern = patterns[::,t].reshape(self.structure[0],1)
            for l in range(L-1):
                if l==0:
                    y = np.vstack((np.array([[1.0]]),pattern))
                else:
                    y = self.local_output[l-1][::,t].reshape(len(self.local_output[l-1][::,t]),1)
                v = np.dot(self.weight[l][::,::,t],y)
                v = v.reshape(len(v),1)
                if l<(L-2):
                    self.local_output[l][0,t] = 1.0
                    for j in range(self.structure[l+1]):
                        self.local_output[l][j+1,t] = self.phi(v[j,0])
                else:
                    for j in range(self.structure[l+1]):
                        self.local_output[l][j,t] = self.phi(v[j,0])
                    
        #calculate the errors
        for t in range(self.TDL):
            for j in range(self.structure[-1]):
                self.output_error[j,t] = targets[j,t] - (self.local_output[-1][j,t]+controls[j,t])
        estimations.append(self.local_output[-1][-1,0]+controls[-1,0])
        
        #record the avg error
        avgE = np.sqrt(np.sum(self.output_error**2)/self.TDL/2.0)
        error_trace.append(avgE)
        return None
    
    def Backward_Pass(self,patterns):
        #calculate the local gradients
        L = len(self.structure)
        
        for t in range(self.TDL):
            for l in np.arange(-1,-1*L,-1):
                if l==-1:
                    for j in range(self.structure[l]):
                        self.local_gradient[l][j,t] = self.output_error[j,t]\
                        *(self.dphi2(self.local_output[l][j,t]))
                elif l==-2:
                    for j in range(self.structure[l]+1):
                        temp = 0.0
                        for k in range(self.structure[l+1]):
                            temp += (self.local_gradient[l+1][k,t])\
                            *(self.weight[l+1][k,j,t])
                        self.local_gradient[l][j,t] = self.dphi2(self.local_output[l][j,t])*temp
                else:
                    for j in range(self.structure[l]+1):
                        temp = 0.0
                        for k in range(self.structure[l+1]):
                            temp += (self.local_gradient[l+1][k+1,t])\
                            *(self.weight[l+1][k,j,t])
                        self.local_gradient[l][j,t] = self.dphi2(self.local_output[l][j,t])*(temp)
                    
        # update the weights
        for l in range(L-1):
            if l==0:
                for j in range(self.structure[l+1]):
                    for i in range(self.structure[l]+1):
                        temp = 0.0
                        for t in range(self.TDL):
                            if i==0:
                                y_i = 1.0
                            else:
                                y_i = patterns[i-1,t]
                            temp += (self.eta)*(self.local_gradient[l][j+1,t])*(y_i)
                        temp = temp/self.TDL
                        self.new_weight[l][j,i] = self.weight[l][j,i,0] + \
                        self.alpha*(self.weight[l][j,i,0]-self.weight[l][j,i,1])+temp
            elif l==(L-2):
                for j in range(self.structure[l+1]):
                    for i in range(self.structure[l]+1):
                        temp = 0.0
                        for t in range(self.TDL):
                            temp+= (self.eta)*(self.local_gradient[l][j,t])*(self.local_output[l-1][i,t])
                        temp = temp/self.TDL
                        self.new_weight[l][j,i] = self.weight[l][j,i,0] + \
                        self.alpha*(self.weight[l][j,i,0]-self.weight[l][j,i,1])+temp
            else:
                for j in range(self.structure[l+1]):
                    for i in range(self.structure[l]+1):
                        temp = 0.0
                        for t in range(self.TDL):
                            temp += (self.eta)*(self.local_gradient[l][j+1,t])*(self.local_output[l-1][i,t])
                        temp = temp/self.TDL
                        self.new_weight[l][j,i] = self.weight[l][j,i,0] + \
                        self.alpha*(self.weight[l][j,i,0]-self.weight[l][j,i,1])+temp

        # shift the weights
        for t in np.arange(-1,-1*(self.TDL+1),-1):
            for l in range(L-1):
                for j in range(self.structure[l+1]):
                    for i in range(self.structure[l]+1):
                        if t == -1*self.TDL:
                            self.weight[l][j,i,t] = self.new_weight[l][j,i]
                        else:
                            self.weight[l][j,i,t] = self.weight[l][j,i,t-1]
        return None
```

下面本报告展示ANN模型在系统辨识中的应用，首先建立如下用于表征ANN模型训练样本的函数。在实际训练ANN模型时候，本报告采用的误差表达式为：
$$E_{\mathrm{avg}}(t)=\frac{1}{T}\sum_{\tau=t-T+1}^{t}{E(\tau)}$$

```python
def Representation_pattern(y, u, t, tdl=4):
    #y should be a 2-D array
    if t<(tdl+1):
        raise("The length of y should be larger than %d"%(tdl+1,))
        
    targets = np.zeros((tdl,),dtype=float)
    patterns = np.zeros((2,tdl),dtype=float)
    controls = np.zeros((tdl,),dtype=float)
    
    for d in range(tdl):
        targets[d] = y[t-d]
        controls[d] = u[t-d-1]
        for j in range(2):
            patterns[j,d] = y[t-d-j-1]
        
    return patterns, targets.reshape(1,tdl), controls.reshape(1,tdl)
```

随后，采用一个2输入1输出、中间2层的ANN模型，ANN的层次结构为：2-20-10-1；学习率采用$\eta=0.25$；动量系数采用$\alpha=0.5$；延迟参数$T=4$；迭代步数设置为600。为了观察结果，本报告绘制出辨识系统的输出结果曲线和误差曲线，具体如下代码所示：

```python
#%% build an ANN model
n = 2
m = 1
structure=(n,20,10,m)
eta=0.25
alpha=0.5
TDL=4
a=1.0

net = ANN(structure, eta, alpha, TDL, a)
net.Initialization()

#train the ANN in an online fashion
maxEpoch = 600
estimations = []
error_trace = []

y = np.zeros((maxEpoch,),dtype=float)
u = np.zeros((maxEpoch,),dtype=float)

for t in range(maxEpoch):
    if t==0:
        u[t] = control(t)
        continue
#    print "Do training at the %d-th epoch......"%(t,)
    if t<=(TDL+1):
        u[t] = control(t)
        y[t] = system_model(y, u[t-1], t)
    else:
        u[t] = control(t)
        y[t] = system_model(y, u[t-1], t)
        patterns, targets, controls = Representation_pattern(y, u, t, TDL)
        net.Forward_Pass(patterns,targets, controls, estimations,error_trace)
        net.Backward_Pass(patterns)
```

实验结果如下图所示：

```python
#%%

epoch = range(maxEpoch)

plt.figure(1,figsize=(8,7))
plt.grid(True)
plt.plot(epoch,y,"-b",lw=4.0,label="Actual $y(t)$")
plt.plot(np.asarray(epoch[-1*(len(estimations))::]),estimations,"-r",lw=4.0,label="Identified $\widehat{y}(t)$")

plt.legend(prop={"size":20.0,"family":"serif"},loc="lower left",
           ncol=2,bbox_to_anchor=(0.04,0.98))
plt.xlabel("Epoch $t$",fontsize=20.0,family="serif")
plt.ylabel("System output",fontsize=20.0,family="serif")
plt.xticks(fontsize=20.0)
plt.yticks(fontsize=20.0)
plt.show()

#%%
plt.figure(2,figsize=(8,7))
plt.grid(True)
plt.plot(np.asarray(epoch[-1*(len(error_trace))::]),error_trace,"-g",lw=4.0,label="MSE")

plt.legend(prop={"size":20.0,"family":"serif"},loc="lower left",
           ncol=1,bbox_to_anchor=(0.04,0.98))
plt.xlabel("Epoch $t$",fontsize=20.0,family="serif")
plt.ylabel("$\sqrt{E_{avg}(t)}$",fontsize=20.0,family="serif")
plt.xticks(fontsize=20.0)
plt.yticks(fontsize=20.0)
plt.show()
```
