在神经网络建模研究领域中，除了广为熟知的BP神经网络模型之外，还有一类十分重要的模型：**径向基神经网络(Radial Basis Function Neural Network, RBF Neural Network)**。RBF神经网络模型是模拟人的视网膜神经元的激励-响应动态特性而提出的数学模型。在生物学上，认为人的视网膜局部区域感受激励信号的强度、并由此产生的响应是与感受视野中心距离相关的：神经元距离感受信号的视野中心距离小，其引起的兴奋程度越大；反之，远离视野中心的神经元响应越小；此外，人们还认为，视网膜神经元的激活区域呈现径向对称。为了刻画这两个生物学特性，在传统的多层感知机模型中引入“径向基函数”作为神经网络的激励函数，而模型的输出是多个径向基输出信号的线性组合 [[1]](#wiki)。RBF神经网络具有很好的非线性逼近能力，已经被广泛应用于函数逼近、时间序列预测、模式分类以及非线性系统控制等领域 [[2]](#ref1), [[3]](#ref2), [[4]](#ref3), [[5]](#ref4)。本次课程报告总结典型的RBF神经网络结构及其数学模型，并以非线性拟合为应用示例展示该神经网络模型的应用。

## 径向基神经网络的数学模型

典型的RBF神经网络模型为一个2层结构的全连接网络，如下图所示[[1]](#wiki):

![RBF network structure](https://upload.wikimedia.org/wikipedia/commons/7/7d/Radial_funktion_network.svg)

RBF神经网络第一层为输入层，第二层为隐层，第三层为输出层。其中，从第一层到第二层的连接权恒为1，而中间隐层神经元的激励函数多采用Gaussian函数。为了方便推导RBF神经网络的数学表达形式，记第$k$个输入信号为列向量$\mathbf{x}_{k}=\left(x_{1,k},x_{2,k},\ldots,x_{m,k}\right)^{\mathrm{T}}$且$\mathbf{x}_{k}\in\mathbb{R}^{m\times1}$，其对应的期望输出信号为$\mathbf{d}_{k}\in\mathbb{R}^{n\times1}$。给定一个由$L$个径向基神经元构成的隐层，输入第$k$个信号$\mathbf{x}_k$，则第$j$个径向基神经元的输出为

$$
\phi_{j}\left(\mathbf{x}_k\right)=\exp\left\{-\beta_j \Vert \mathbf{x}_k - \mathbf{c}_j\Vert_{2}^{2}\right\}
$$

记中间隐层中第$j$个径向基神经元到输出层中第$i$个神经元的连接权为$w_{i,j}$，偏置为$b_i$，则该神经元的输出为$y_{i,k}$，可以得到：

$$
y_{i,k}=\sum_{j=1}^{L}{w_{i,j}\phi_{j}\left(\mathbf{x}_k\right)} + b_i
$$
其中，$i=1,2,\ldots,n$。

给定$\mathbf{y}_k=\left(y_{1,k},y_{2,k},\ldots,y_{n,k}\right)^{\mathrm{T}}\in\mathbb{R}^{n\times 1}$，$\mathbf{b}=\left(b_1,b_2,\ldots,b_n\right)^{\mathrm{T}}\in\mathbb{R}^{n\times 1}$，以及$\mathbf{W}=\left[w_{i,j}\right]\in\mathbb{R}^{n\times L}$，即可以得到RBF神经网络各层输出的计算模型：

$$
\begin{cases}
\mathbf{z}\left(\mathbf{x}_k\right)=\left(\phi_1(\mathbf{x}_k),\phi_{2}(\mathbf{x}_k),\ldots,\phi_{L}(\mathbf{x}_k)\right)^{\mathrm{T}},&\mathbf{z}\left(\mathbf{x}_k\right)\in\mathbb{R}^{L\times 1};\\
\mathbf{y}_k = \mathbf{W}\mathbf{z}\left(\mathbf{x}_k\right)+\mathbf{b}；
\end{cases}
$$
其中，$k=1,2,\ldots,N$。

值得一提的是，在给定训练样本集合的前提下，RBF神经网络模型可以表达为另一种更为简洁的矩阵方程形式：

$$
\mathbf{Y}=\mathbf{W}\mathbf{Z}\left(\mathbf{X}\right)+\mathbf{b}
$$
其中，$\mathbf{Y}=\left[\mathbf{y}_1,\mathbf{y}_2,\ldots,\mathbf{y}_N\right]\in\mathbb{R}^{n\times N}$，$\mathbf{Z}\left(\mathbf{X}\right)=\left[\mathbf{z}\left(\mathbf{x}_1\right),\mathbf{z}\left(\mathbf{x}_2\right),\ldots,\mathbf{z}\left(\mathbf{x}_N\right)\right]\in\mathbb{R}^{L\times N}$，以及$\mathbf{X}=\left[\mathbf{x}_1,\mathbf{x}_2,\ldots,\mathbf{x}_N\right]\in\mathbb{R}^{m\times N}$。

此外，中间隐层的径向基神经元输出信号可以进一步"归一化"从而得到具有更好鲁棒性的RBF神经网络：

$$
\phi_j\left(\mathbf{x}_k\right)\leftarrow \frac{\phi_j\left(\mathbf{x}_k\right)}
{\sum_{j=1}^{L}{\phi_j\left(\mathbf{x}_k\right)}} 
$$

## 径向基神经网络的学习算法

在给定一个有限的训练样本集合以及RBF神经网络拓扑结构前提下，确定该RBF神经网络则需要确定三类参数：网络中间隐层各个径向基神经元的中心向量：$\mathbf{c}_j$，标量系数$\beta_j$，以及中间隐层和输出层神经元的连接权$w_{i,j}$和偏置$b_i$ ($j=1,\ldots,L$; $i=1,\ldots,n$)。典型的学习算法有三类：

+ 聚类算法和梯度训练算法相结合的学习策略：首先，通过聚类算法确定中间隐层径向基神经元的中心向量和标量系数；然后，利用梯度训练算法，利用给定的训练样本迭代确定中间层和输出层之间的连接权和偏置；
+ 梯度训练策略：利用给定的训练样本，采用梯度下降训练迭代确定中间隐层径向基神经元的中心向量和标量系数，以及中间层和输出层之间的连接权和偏置；
+ 聚类算法和求解矩阵广义逆相结合的算法：同样先通过聚类算法确定中间隐层径向基神经元的中心向量和标量系数；然后，利用求解矩阵广义逆的方法确定中间层和输出层之间的连接权和偏置（将偏置当成一种特殊的连接权，其对应的输入信号为1）。其中，矩阵广义逆的求解算法需要依赖于正交最小二乘算法。

在本次报告中，将采用“聚类算法和梯度训练算法相结合的学习策略”确定RBF神经网络模型的参数。具体的学习策略如下步骤：

1. 首先，利用给定的训练样本集合以及RBF神经网络的拓扑结构($m-L-n$)，采用**DBSCAN聚类算法**确定该样本集合的$L$个聚类中心：$\mathbf{c}_j$ ($j=1,\ldots,L$)，作为中间隐层$L$个径向基神经元的中心向量；
2. 其次，记以$\mathbf{c}_j$为中心的“类”为集合$\mathbf{C}_j$，样本$\mathbf{x}$属于类$\mathbf{C}_j$记为$\mathbf{x}\in\mathbf{x}_j$,根据Gaussian正态分布的公式形式则可通过如下式子确定第$j$个径向基神经元的标量系数$\beta_j$：
$$
\beta_j = \frac{1}{2\sigma_j^2}
$$
其中,$\sigma_j$为：
$$
\sigma_j=\frac{1}{\vert \mathbf{C}_j\vert}
\sum_{\mathbf{x}\in\mathbf{C}_j}\Vert \mathbf{x}-\mathbf{c}_j\Vert_2
$$
3. 然后，根据梯度训练算法迭代确定中间层和输出层之间的连接权和偏置。需要特地指出，为了提高RBF神经网络的泛化能力，本报告的目标函数式子引入“正则化”项：
$$
H = \frac{1}{2N}\sum_{k=1}^{N}{\Vert \mathbf{y}_k-\mathbf{d}_k\Vert}_{2}^{2}+\frac{\lambda}{2N}\sum_{j=1}^{L}{\sum_{i=1}^{n}{w_{i,j}^2}}
$$
据此，该目标函数式子对应于权值$w_{i,j}$和偏置$b_i$的梯度分别为：
$$
\begin{cases}
\frac{\partial H}{\partial w_{i,j}} = \frac{1}{N}\sum_{k=1}^{N}(y_{i,k}-d_{i,k})\phi_{j}(\mathbf{x}_k)+\frac{\lambda}{N}w_{i,j}\\
\frac{\partial H}{\partial b_i} = \frac{1}{N}\sum_{k=1}^{N}(y_{i,k}-d_{i,k})+\frac{\lambda}{N}w_{i,j}
\end{cases}
$$
给定第$t$次迭代中的权值和偏置分别为$w_{i,j}(t)$和$b_i(t)$，则有
$$
\begin{cases}
w_{i,j}(t+1)=\left(1-\frac{\eta\lambda}{N}\right)w_{i,j}(t) - \frac{\eta}{N}\sum_{k=1}^{N}{(y_{i,k}-d_{i,k})\phi_{j}(\mathbf{x}_k)},~i=1,\ldots,n;~j=1,\ldots,L;\\
b_{i}(t+1)=b_{i}(t) - \frac{\eta}{N}\sum_{k=1}^{N}{(y_{i,k}-d_{i,k})},~i=1,\ldots,n.
\end{cases}
$$

在训练样本量十分巨大的情况下，亦即$N$比较大，上述梯度训练算法进行一次迭代耗费的时间将会很长。这是因为，梯度更新更公式依赖于计算$N$个样本的误差$(y_{i,k}-d_{i,k})$。为了加快迭代速度，本报告采用“**随机梯度下降**”（**Stochastic Gradient Descent，SDG**）训练算法训练RBF神经网络。其中，进行每一次迭代，随机从样本集合中选取$M$个样本训练神经网络，且$M<<N$。因此，实际应用中，SGD训练算法采用如下迭代更新公式：

$$
\begin{cases}
w_{i,j}(t+1)=\left(1-\frac{\eta\lambda}{N}\right)w_{i,j}(t) - \frac{\eta}{M}\sum_{k=1}^{M}{(y_{i,r_k}-d_{i,r_k})\phi_{j}(\mathbf{x}_{r_k})},~i=1,\ldots,n;~j=1,\ldots,L;\\
b_{i}(t+1)=b_{i}(t) - \frac{\eta}{M}\sum_{k=1}^{M}{(y_{i,r_k}-d_{i,r_k})},~i=1,\ldots,n.
\end{cases}
$$
其中，$\left\{r_1,r_2,\ldots,r_M\right\}\subsetneq\left\{k,k=1,2,\ldots,N\right\}$。

## 径向基神经网络的实现

本报告中采用“**DBSCAN聚类算法**”+“**SDG算法**”实现RBF神经网络的学习过程，并以非线性拟合作为具体问题展示该神经网络模型的应用，所要逼近的非线性函数为如下二元非线性多峰值函数：

$$
f(x,y) = 3(1-x)^2e^{-x^2-(y+1)^2}-10\left(\frac{x}{5}-x^3-y^5\right)e^{-x^2-y^2}-\frac{1}{3}e^{-(x+1)^2-y^2}
$$

其中，DBSCAN聚类算法(Density-Based Spatial Clustering of Applications with Noise)是一种非监督式的、基于样本密度分布特征的聚类算法，能够识别噪声点儿具有较好的分类鲁棒性。由于在给定类的最大半径和类的最小邻域数目条件下，DBSCAN能够自动对数据样本进行“归类”、自动确定“类”的数目。据此，本报告利用DBSCAN聚类算法确定RBF神经网络训练样本的“类”数目、根据该数目确定RBF层神经元数目。

RBF神经网络的实现Python代码具体如下：

```python
%matplotlib inline

# Import the third-party modules
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import random


# Global plot properties
fontdict = {"fontsize": 20.,
           "family": "serif"}

# Define a nonlinear 2-D function
def f(x,y):
    s1 = 3.0*(1.0-x)**2.0*np.exp(-x**2.0-(y+1.0)**2.0)
    s2 = -10.0*(x/5.0-x**3.0-y**5.0)*np.exp(-x**2.0-y**2.0)
    s3 = -(1.0/3.0)*(np.exp(-(x+1.0)**2.0-y**2.0))
    return s1+s2+s3


# Show the true map of the function
xmin, xmax = -2., 2.
ymin, ymax = -2., 2.
xnum = 500
ynum = 500
X, Y = np.meshgrid(np.linspace(xmin,xmax,xnum),np.linspace(ymin,ymax,ynum), sparse=False, indexing='xy')
Z = f(X,Y)
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure(0,figsize=(10,8))
ax = fig.gca(projection='3d')

ax.plot_surface(X, Y, Z, rstride=10, cstride=10, alpha=0.2)
cset = ax.contourf(X, Y, Z, zdir='z', offset=-Z.max(), cmap=cm.jet)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-xmax, cmap=cm.jet)
cset = ax.contourf(X, Y, Z, zdir='y', offset=ymax, cmap=cm.jet)

ax.set_xlabel('X',fontdict=fontdict)
ax.set_ylabel('Y',fontdict=fontdict)
ax.set_zlabel('Z',fontdict=fontdict)
ax.set_xlim(xmin-.5, xmax+.5)
ax.set_ylim(ymin-.5, ymax+.5)
ax.set_zlim(Z.min()-.5, Z.max()+.5)

plt.show()
```

```python
# Define sigmoid function
def sigmoid(x):
    return 1.0/(1.0+np.exp(-x))

# Define the derivative of the sigmoid function
def dsigmoid(x):
    return sigmoid(x)*(1.0-sigmoid(x))

# The cost function 
def Cost_function(a, y):
    return 0.5*np.linalg.norm(a-y)

# The error delata of the output layer
# Given that the output layer adopts the weighted linear equation
def delta(z, a, y):
    return (a-y)*1.0


# Define the RBF
def RBF(x, beta, c):
    return np.exp(-beta*np.linalg.norm(x-c))

# Define the RBF network
class RBF_Network(object):
    
    def __init__(self, structure):
        """structure is a list containing the number of each layer in the network"""
        self.layer_num = len(structure)
        self.structure = structure
        if self.layer_num>3:
            print "The total number of layers is more than 3!"
            print "Except for the first two layers (the input and the RBF layer) and the output layer, "
            print "the activation function of each neuron in these hidden layers is simulated with a Sigmoid function."
        elif self.layer_num<3:
            raise("The total number of layers should not be less than 3!!")
    
    
    def initialize_weights_and_biases(self):
        # Initialize the weights and the biases between the hidden layer(s) and the output layer
        mlp_size = self.structure[1:]
        self.b = [np.random.randn(m, 1) for m in mlp_size[1:]]
        self.w = [np.random.randn(m,L)/np.sqrt(L) for m, L in zip(mlp_size[1:], mlp_size[:-1])]
        
    
    def update_RBF_layer(self, training_data, show_clusters=False):
        """The 'training_data' is a list of tuples (x,y) where 'x' is the input column vector
        and 'y' is the output column vector. Here, the DBSCAN algorithm is performed to determine
        the prototype of each RBF neuron in the RBF layer."""
        
        total_data_num = len(training_data)
        eps = ((total_data_num-1000)/(2000-1000))*(0.15-0.2)+0.2
        
        from sklearn.cluster import DBSCAN
        
        estimator = DBSCAN(eps=eps, min_samples=10)

        X = np.zeros((total_data_num,self.structure[0]), dtype=float)
        flag = 0
        for (x,y) in training_data:
            X[flag,::] = x.reshape((2,))
            flag += 1
        
        # fit the DBSCAN classificator
        # print X[0:3,::]
        estimator.fit(X)
        labels = estimator.labels_
        
        n_clusters = len(np.unique(labels)) - (1 if -1 in labels else 0)
        print "Totally {} clusters are found by DBSCAN algorithm!".format(n_clusters)
        if n_clusters != self.structure[1]:
            print "Now, the number of RBF neurons is changed from {0} to {1}".format(self.structure[1], n_clusters)
            self.structure[1] = n_clusters
        
        print "Set the prototypes of RBF neurons to be the centriods of clusters, respectively."
        unique_labels = set(labels)
        prototypes = np.zeros((n_clusters,len(X[0,::])), dtype=float)
        scalars = np.zeros((n_clusters,),dtype=float)
        cflag = 0
        print "The center vectors of RBF neurons are:"
        for k in unique_labels:
            if k != -1:
                # except for the noise data
                class_member_mask = (labels == k)
                members = X[class_member_mask]
                prototypes[cflag,::] = np.mean(members,0)
                print "The {0}-th center vector is ({1},{2})".format(cflag,prototypes[cflag,0],prototypes[cflag,1])
                scalars[cflag] = np.sum([np.linalg.norm(point-np.mean(members,0)) for point in members])/len(members)
                scalars[cflag] = 1.0/(2.0*((scalars[cflag])**2.0))
                print "The scalar is {}".format(scalars[cflag])
                cflag += 1
        
        # Initialize the prototype of each RBF neuron and its scalar coefficient
        self.c = prototypes
        self.beta = scalars
        
        if show_clusters:
            plt.figure(1,figsize=(10,8))
            core_samples_mask = np.zeros_like(estimator.labels_, dtype=bool)
            core_samples_mask[estimator.core_sample_indices_] = True
            colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
            for k, col in zip(unique_labels, colors):
                if k == -1:
                    # Black used for noise.
                    col = 'k'

                class_member_mask = (labels == k)

                xy = X[class_member_mask & core_samples_mask]
                plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                         markeredgecolor='k', markersize=14)

                xy = X[class_member_mask & ~core_samples_mask]
                plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                         markeredgecolor='k', markersize=6)

            plt.title('Estimated number of clusters: %d' % n_clusters, fontsize=20.)
            plt.show()
            
            
    def forward(self, x):
        # Evaluate the output of each RBF neuron in the RBF layer
        a = np.zeros((self.structure[1],1),dtype=float)
        for l in range(self.structure[1]):
            mu = self.c[l,::]
            a[l,0] = RBF(x, self.beta[l], mu.reshape(x.shape))
            
        # Normalize the output of each RBF neuron
        a = a/np.sum(a)
        
        for w, b in zip(self.w[:-1], self.b[:-1]):
            a = sigmoid(np.dot(w,a)+b)
        w = self.w[-1]
        b = self.b[-1]
        a = np.dot(w,a)+b # the output layer adopts the weighted linear equation
        
        return a
    
    def backward_propagation(self, x, y):
        nb = [np.zeros(b.shape) for b in self.b]
        nw = [np.zeros(w.shape) for w in self.w]
        
        # calculate the output of the RBF layer
        activation = np.zeros((self.structure[1],1),dtype=float)
        for l in range(self.structure[1]):
            mu = self.c[l,::]
            activation[l,0] = RBF(x, self.beta[l], mu.reshape(x.shape))
            
        # forward pass
        activation = activation/np.sum(activation)
        activations = [activation]
        outputs = []
        for w, b in zip(self.w[:-1],self.b[-1]):
            output = np.dot(w,activation)+b
            outputs.append(output)
            activation = sigmoid(output)
            activations.append(activation)
        w = self.w[-1]
        b = self.b[-1]
        output = np.dot(w,activation)+b
        outputs.append(output)
        activation = output
        activations.append(activation)
        
        # backward pass
        delt = delta(outputs[-1],activations[-1],y)
        nb[-1] = delt
        nw[-1] = np.dot(delt, activations[-2].transpose())
        
        for l in xrange(2,self.layer_num-1):
            output = outputs[-l]
            sp = dsigmoid(output)
            delt = np.dot(self.w[-l+1].transpose(),delt)*sp
            nb[-l] = delt
            nw[-l] = np.dot(delt, activations[-l-1].transpose())
        return (nb, nw)
    
    
    def update_batch_SGD(self, batch, eta, lmbda, N):
        nb = [np.zeros(b.shape) for b in self.b]
        nw = [np.zeros(w.shape) for w in self.w]
        for (x, y) in batch:
            delta_nb, delta_nw = self.backward_propagation(x,y)
            nb = [b+db for b, db in zip(nb, delta_nb)]
            nw = [w+dw for w, dw in zip(nw, delta_nw)]
        
        # Gradient descent
        self.b = [b-(eta/len(batch))*db for b, db in zip(self.b, nb)]
        self.w = [(1.0-eta*(lmbda/N))*w-(eta/len(batch))*dw for w, dw in zip(self.w, nw)]
        
    
    def training(self, training_data, epochs, batch_size, eta, lmbda=0.0):
        N = len(training_data)
        cost = []
        accuracy = []
        for j in xrange(epochs):
            random.shuffle(training_data)
            batches  = [training_data[l:l+batch_size] for l in xrange(0,N,batch_size)]
            for batch in batches:
                self.update_batch_SGD(batch, eta, lmbda, N)
            if j%50==0: print "Complete the {0}-th epoch training...".format(j)
            
            # Record the training trace
            cost.append(self.cost_function(training_data, lmbda))
            accuracy.append(self.evaluate_accuracy(training_data))
            
        return cost, accuracy
    
    def cost_function(self, training_data, lmbda):
        cost = 0.0
        for x, y in training_data:
            output = self.forward(x)
            cost += Cost_function(output,y)/len(training_data)
        cost += 0.5*(lmbda/len(training_data))*sum(np.linalg.norm(w)**2.0 for w in self.w)
        return cost
    
    def evaluate_accuracy(self, data):
        ac = 0.0
        for x, y in data:
            o = self.forward(x)
            ac += Cost_function(o,y)/len(data)
        return ac

def format_training_data(x,y,z):
    training_data = [(np.array([[xi],[yi]]),np.array([[zi]])) for xi,yi,zi in zip(x,y,z)]
    print "The point number is {}".format(len(x))
    print "The shape of each input is (%d,%d)"%training_data[0][0].shape
    print "The shape of each output is (%d,%d)"%training_data[0][1].shape
    return training_data
```

```python
# Hold the random seeder
np.random.seed(0)
point_num = 2000 # sample number
# Initialize the input-output pairs [(xi,yi),i=1,...,N]
x = np.random.uniform(xmin, xmax, point_num)
y = np.random.uniform(ymin, ymax, point_num)
z = f(x,y)

training_data = format_training_data(x,y,z)

# Set up the hyper parameters of a RBF network
structure = [2,30,1]
net = RBF_Network(structure)
# Determine the prototypes and the scalars of RBF neurons in the RBF layer
net.update_RBF_layer(training_data, show_clusters=True)
```

```python
# Traing the RBF network w.r.t. weights and biases
epochs = 1000
batch_size = 100
eta = 0.2
lmbda = 0.1

net.initialize_weights_and_biases()
(cost, accuracy) = net.training(training_data, epochs, batch_size, eta, lmbda)

# Show the training result
plt.figure(2,figsize=(10,8))
plt.subplot(2, 1, 1)
plt.grid(True)
plt.plot(np.arange(epochs), cost, 'y-', lw=4.0)
plt.ylabel('Cost $H$',fontdict=fontdict)

plt.subplot(2, 1, 2)
plt.grid(True)
plt.plot(np.arange(epochs), accuracy, 'r-', lw=4.0)
plt.xlabel('Epochs',fontdict=fontdict)
plt.ylabel('Accuracy',fontdict=fontdict)

plt.show()
```

# 总结

本次课程报告给出了径向基神经网络的数学原理和基本的学习算法。通过非线性的二元函数拟合作为应用示例，展示径向基神经网络在函数逼近中的应用途径。其中，从实际的数值实验可知，所构建的径向基神经网络在复杂的非线性函数逼近过程中，径向基神经网络能够大体上近似给出的函数。然而，值得指出的是，本次报告仅出于应用示例的目的进行数值实验，实验中没有对径向基神经网络涉及的超参数以及聚类算法涉及的参数等相关设置进行优化（调优），因此，实验得到的网络逼近效果和真实函数仍然具有一定差距。值得指出，**由于神经网络的收敛性能受到聚类算法所确定的径向基中心向量$\mathbf{c}_j$和标量系数$\beta_j$、径向基神经元个数(聚类算法确定的聚类中心数目)、梯度算法的学习速率以及动量系数等多个参数的综合的影响，想要在实际复杂函数逼近应用中使径向基神经网络收敛于较好的逼近性能，则需要投入较大的精力和经验对这些参数进行联合调优**，此外，由于径向基神经网络的径向基神经元层的参数确定除了可以使用聚类算法，还可以采用梯度算法进行确定。这也说明，在实际工程应用中，神经网络模型的应用需要根据问题的实际背景而展开，作为一种典型的”黑箱“模型，其推广应用并不那么直接，需要结合参数调优的经验、训练样本的选取等条件。
